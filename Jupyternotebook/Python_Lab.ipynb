```python
class py_solution:
    def int_to_Roman(self, num):
        val = [
            1000, 900, 500, 400,
            100, 90, 50, 40,
            10, 9, 5, 4,
            1
            ]
        syb = [
            "M", "CM", "D", "CD",
            "C", "XC", "L", "XL",
            "X", "IX", "V", "IV",
            "I"
            ]
        roman_num = ''
        i = 0
        while  num > 0:
            for _ in range(num // val[i]):
                roman_num += syb[i]
                num -= val[i]
            i += 1
        return roman_num


print(py_solution().int_to_Roman(1))
print(py_solution().int_to_Roman(4000))
```

```python
def list_of_dicts(marks):
    keys = marks.keys()
    vals = marks[k]
    zip(*[ for k in keys])
    result = [dict(zip(keys, v)) for v in vals]
    return result

marks = {'Science': [88, 89, 62, 95], 'Language': [77, 78, 84, 80]}
print("Original dictionary of lists:")
print(marks)
print("\nSplit said dictionary of lists into list of dictionaries:")
print(list_of_dicts(marks))
```

```python
def removeDuplicate(str, n):
      
   
    index = 0
      
    
    for i in range(0, n):
          
        
        for j in range(0, i + 1):
            if (str[i] == str[j]):
                break
                  
        
        if (j == i):
            str[index] = str[i]
            index += 1
              
    return "".join(str[:index])
  

str= "Chandigarh".lower()

n = len(str)
print(removeDuplicate(list(str), n))

str= "Madam".lower()
n = len(str)
print(removeDuplicate(list(str), n))
```

```python

def frequency(my_list):
    
    freq = {}
    for item in my_list:
        if (item not in freq):
            freq[item] = 0
        freq[item] += 1

    for key, value in freq.items():
        print ("% d : % d"%(key, value))
L=[2,4,2,6,3,1,6,2] 
frequency(L)
```
