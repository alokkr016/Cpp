```python
def my_mean(l):
    n = len(l)
    total_sum = sum(l)
    mean_my = total_sum/n
    print("Mean is " + str(mean_my))
def my_meadian(l):
    n = len(l)
    l.sort()
    if n % 2 == 0:
        median1 = l[n//2]
        median2 = l[n//2 - 1]
        median = (median1 + median2)/2
    else:
        median = l[n//2]
    print("Median is: " + str(median))

def my_mode(y):
    y.sort()
    L1=[]

    i = 0
    while i < len(y) :
    	L1.append(y.count(y[i]))
    	i += 1

    d1 = dict(zip(y, L1))
    d2={k for (k,v) in d1.items() if v == max(L1) }
    print("Mode(s) is/are :" + str(d2))   
        

my_num = [3,7,5,8,4,4,4,3,1,2,6,7]
while(True): 
    print("What do you want to do ")
    print("Press 1 for mean \nPress 2 for median\nPress 2 for Mode\nPress 4 to exit the program")
    value = input("Enter your choice ")
  
    if value == '1':
        my_mean(my_num)
    if value == '2':
        my_meadian(my_num)
    if value == '3':
        my_mode(my_num)
    if value == '4':
        exit()
```

```python
SIZE=10
stack=[0 for x in range(SIZE)]
stp=-1
def push(x):
    global stp
    global stack
    stp=stp+1
    stack[stp]=x

def pop():
    global stp
    global stack
    x=stack[stp]
    stp=stp-1
    return x

def is_stack_empty():
    global stp
    return stp==-1

def is_stack_full():
    global stp
    return stp==SIZE-1

def disp_stack():
    global stp
    global stack
    for x in range(stp+1):
        print(stack[x])


while (1):
    choice=int(input("enter choice: 1 for push, 2 for pop, 3 for display"))
    if choice==1:
        if is_stack_full():
            print("Sorry stack is full")
        else:
            x=int(input("enter the value to be pushed:"))
            push(x)
    elif choice==2:
        if is_stack_empty():
            print("Sorry stack is empty")
        else:
            x=pop()
            print("popped value is:",x)
    elif choice==3:
        if is_stack_empty():
            print("Sorry stack is empty")
        else:
            disp_stack()
    elif choice==4:
        break
```

```python
def mean1():
    n_num =  [ 1, 2, 3, 4, 5]
    n = len( n_num)

    get_sum =  sum(n_num)
    mean =  get_sum / n

    print("Mean / Average is: " +  str(mean)) 

def median1():
    n_num = [1, 2, 3, 4, 5]
    n = len(n_num)
    n_num.sort()
    
    if n % 2 == 0:
        median1 = n_num[n//2]
        median2 = n_num[n//2 - 1]
        median = (median1 + median2)/2
    else:
        median = n_num[n//2]
    print("Median is: " + str(median))


def mode1():
    from collections import Counter


    n_num = [1, 2, 3, 4, 5, 5]
    n = len(n_num)

    data = Counter(n_num)
    get_mode = dict(data)
    mode = [k for k, v in get_mode.items() if v == max(list(data.values()))]

    if len(mode) == n:
    	get_mode = "No mode found"
    else:
    	get_mode = "Mode is / are: " + ', '.join(map(str, mode))
    
    print(get_mode)
while(True):
    print("What do you want to do ")
    print("Press 1 for mean \nPress 2 for median\nPress 2 for Mode\nPress 4 to exit the program")
    value = input("Enter your choice ")
    if value == '1':
        mean1()
    if value == '2':
        median1()
    if value == '3':
        mode1()
    if value == '4':
        exit()
```
